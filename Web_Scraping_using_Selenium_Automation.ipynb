```python
url = "https://www.imdb.com/chart/moviemeter/?ref_=nv_mv_mpm"
```

```python
requests.get(url) #Get Request to url
```

```python
headers = {"User-Agent":"Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36"}
```

```python
source = requests.get(url,headers = headers)
source
```

```python
soup = BeautifulSoup(source.content,'lxml') #Create Soup
```

```python
#Create Empty list
Rank = []
Movie_name = []
Year = []
Rating = []
Votes = []
```

```python
Movie_data = soup.findAll('div',attrs ={'class':"ipc-metadata-list-summary-item__c"}) #All movie divs
```

```python
for i in Movie_data:
    rank = i.find('div',class_ = 'sc-b8b74125-0 eukXAN meter-const-ranking sc-be6f1408-6 dGVRjf cli-meter-title-header').text
    rank = re.sub(r'\(([^)]*)\)','',rank) #Remove Text in Parentheses r'\([^)]*\)
    Rank.append(rank)

    movie = i.find('h3',class_ = "ipc-title__text")
    Movie_name.append(movie)

    value = i.find_all('span',class_ = "sc-be6f1408-8 fcCUPU cli-title-metadata-item")
    year = value[0].text
    Year.append(year)

    rating = i.find('span',class_="sc-be6f1408-1 dbnleL").text.replace('Rate','')
    rating = re.sub(r'\([^)]*\)','',rating)
    Rating.append(rating)

    vote = i.find('span',class_ = "sc-be6f1408-1 dbnleL").text.replace('Rate','')
    vote = re.findall(r'\(([^)]*)\)', vote) #Extract text in parenthes
    Votes.append(vote)
```

```python
import pandas as pd

# Ensure all lists have the same length by padding missing values
max_length = max(len(Rank), len(Movie_name), len(Year), len(Rating), len(Votes))
Rank.extend([''] * (max_length - len(Rank)))
Movie_name.extend([''] * (max_length - len(Movie_name)))
Year.extend([''] * (max_length - len(Year)))
Rating.extend([''] * (max_length - len(Rating)))
Votes.extend([''] * (max_length - len(Votes)))

# Create DataFrame
df = pd.DataFrame({'Rank': Rank, 'Name of Movie': Movie_name, 'Year of release': Year, 'Ratings': Rating, 'Votes': Votes})
df
```
